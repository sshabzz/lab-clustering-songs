```python
import pandas as pd
import numpy as np
```

```python
#importing the dataframes
hot100 = pd.read_csv("hot100_features.csv")
not_hot = pd.read_csv("not_hot_features.csv")
```

```python
hot100.head()
```

```python
not_hot.head()
```

```python
#dropping duplicates in not_hot dataframe
not_hot.drop_duplicates(subset=["id"],keep="first", inplace=True, ignore_index=True)
```

```python
not_hot.shape
```

```python
#addint "hot" column to the dataframe
hot = ["Yes"]*100
hot100["hot"] = hot
hot100
```

```python
hot = ["No"]*2845
not_hot["hot"] = hot
not_hot
```

```python
#concatinating 2 dataframes
all_songs = pd.concat([hot100, not_hot], axis=0)
all_songs.shape
all_songs.head()
```

```python
#importing everything for the clustering
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
all_songs.dtypes
```

```python
#dropping the columns that we won´t use for clustering:
X = all_songs.select_dtypes(include=np.number).drop(columns=["key", "mode", "duration_ms", "time_signature"])
X
```

```python
X.describe()
```

```python
#Scaling the data and saving the scaler:
import pickle

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
filename = "scaler.pickle" # Path with filename

#saving the scaler:
with open(filename, "wb") as file:
        pickle.dump(scaler,file)

X_scaled_df = pd.DataFrame(X_scaled, columns = X.columns)
display(X.head()) #before scaling
print()
display(X_scaled_df.head())
```

### KMeans

```python
#testing #k from 10 to 19:

K = range(10, 19)

inertia = []
silhouette = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                    verbose=1)
    kmeans.fit(X_scaled_df)
    
    filename = "kmeans_" + str(k) + ".pickle" # Path with filename
    with open(filename, "wb") as file:
        pickle.dump(kmeans,file)

    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X_scaled_df, kmeans.predict(X_scaled_df)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(K, inertia, 'bx-')
ax[0].set_xlabel('k')
ax[0].set_ylabel('inertia')
ax[0].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[0].set_title('Elbow Method showing the optimal k')
ax[1].plot(K, silhouette, 'bx-')
ax[1].set_xlabel('k')
ax[1].set_ylabel('silhouette score')
ax[1].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[1].set_title('Silhouette Method showing the optimal k')
```

```python
#based on Silhouette Method, we think that K=14, K=16 and K=17 could be the options for clustering:
```

```python
kmeans = KMeans(n_clusters=14, random_state=1234)
kmeans.fit(X_scaled_df)
```

```python
clusters = kmeans.predict(X_scaled_df)
pd.Series(clusters).value_counts().sort_index()
```

```python
# checking the model with k=16:
kmeans = KMeans(n_clusters=16, random_state=1234)
kmeans.fit(X_scaled_df)

clusters1 = kmeans.predict(X_scaled_df)
pd.Series(clusters1).value_counts().sort_index()
```

```python
# checking the model with k=17:
kmeans = KMeans(n_clusters=17, random_state=1234)
kmeans.fit(X_scaled_df)

clusters2 = kmeans.predict(X_scaled_df)
pd.Series(clusters2).value_counts().sort_index()
```

### DBSCAN

```python
from sklearn.datasets import make_classification
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture
from scipy.spatial import distance_matrix
```

```python
# After testing various eps, we saw that the model prefers to create one big cluster and some noise. 
# This is not an option for us, therefore we will not use DBSCAN in our recommender.
model = DBSCAN(eps=4.5, min_samples=18)
yhat = model.fit_predict(X_scaled_df)
print("The Silhouette score of the model is: ",round(silhouette_score(X_scaled_df, model.fit_predict(X_scaled_df)),2))
```

```python
yhat
```

```python
X['cluster'] = yhat
X["cluster"].value_counts()
```

```python
clusters = np.unique(yhat)
clusters
```

```python
#Step1
d = distance_matrix(X_scaled_df,X_scaled_df)
print(d)
print()
```

```python
#Step2
d2 = np.sort(d)[:,1:]
print(d2)
```

```python
#Step3
f = d2[:,17]
print(f)
```

```python
#Step4
f_final = np.sort(f)[::-1]
print()
print(f_final)
```

```python
#Step5
plt.scatter(range(1,len(f_final)+1), f_final)
plt.xlabel("Index") #plotting vertical line
plt.ylabel("Distance") #plotting horizontal line
# Get the index of the distance which is closer to 0.5 as 0.5 is the default value for epsilon in DBSCAN
min_index = np.argmin(np.abs(f_final - 0.5))
xnew = min_index
ynew = f_final[min_index]
plt.axvline(xnew, color='red', linewidth=0.5)
plt.axhline(ynew, color='red', linewidth=0.5)
plt.grid(b=True)
plt.show()
```

```python
np.arange(1.8,4,0.05)
```

```python
models = {}

for value in np.arange(1.8,4,0.1):
    model = DBSCAN(eps=value, min_samples=18) # min_samples = 2 * #features
    # fit model and predict clusters
    yhat = model.fit_predict(X_scaled_df)
    score =  round(silhouette_score(X_scaled_df, model.fit_predict(X_scaled_df)),2)
    models["model_"+str(round(value,2))] = {"model":model,"yhat": yhat, "score": score}
```

```python
# retrieve unique clusters of one model
clusters2 = np.unique(models["model_2.0"]["yhat"])
clusters2
```

```python
models["model_2.0"]["score"]
```

```python
import matplotlib.ticker as ticker

epsilons = [round(value,2) for value in np.arange(1.8,4,0.1)]
scores = [item[1]["score"] for item in list(models.items())]

fig, ax = plt.subplots(figsize=(12,7))
ax.plot(epsilons, scores)
ax.set_xlabel("Epsilon")
ax.set_ylabel("Silhouette score")
ax.xaxis.set_ticks(np.arange(min(epsilons), max(epsilons), 0.1))
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%0.2f'))
ax.yaxis.set_ticks(np.arange(min(scores), max(scores)+0.1, 0.02))
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.2f'))
plt.grid()
plt.tight_layout()
plt.show()
```
